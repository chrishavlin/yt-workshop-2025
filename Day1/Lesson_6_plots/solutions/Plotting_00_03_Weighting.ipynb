## On the importance of weigting fields 

```python
import yt 
ds = yt.load_sample("IsolatedGalaxy")
```

Create a plot that shows the number of grid cells along the x axis:
* use a `ProjectionPlot`
* use the special `('index', 'ones')` field with no weight field
* use a raw summation (`method='sum') 

```python
slc = yt.ProjectionPlot(ds, 'x', ('index', 'ones'), 
                        weight_field=None, 
                        method='sum')
slc.annotate_grids()
slc.set_log(('index', 'ones'), False)
slc.show()
```

Create a standard unweighted histogram using a PhasePlot

```python
yt.PhasePlot(ds.all_data(), 
             ('gas', 'mass'), 
             ('gas', 'temperature'), 
             ('index', 'ones'),
              weight_field=None)
```

```python
yt.PhasePlot(ds.all_data(), 
             ('gas', 'density'), 
             ('gas', 'temperature'), 
             ('index', 'ones'),
              weight_field=None)
```

```python
yt.PhasePlot(ds.all_data(), 
             ('gas', 'density'), 
             ('gas', 'temperature'), 
             ('index', 'cell_volume'),
              weight_field=None)
```
